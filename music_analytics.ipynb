```python
import pandas as pd
import numpy as np
```

```python
songs_dataset = pd.read_json('MasterSongList.json')
```

```python
songs_dataset.head(5)
```

```python
test_split= ["thanh:nguyen"]
test_split[0].split(':')[0]
```

```python
songs_dataset.loc[:,'genres'] = songs_dataset['genres'].apply(''.join)
def consolidateGenre(genre):
    if(len(genre) > 0):
        return genre.split(':')[0]
    else:
        return genre
songs_dataset.loc[:,'genres'] = songs_dataset['genres'].apply(consolidateGenre)
```

```python
audio_feature_list = []
for audio_feature in songs_dataset['audio_features']:
    audio_feature_list.append(audio_feature)
```

```python
audio_feature_header = ['key','energy','liveliness','tempo','speechiness','acousticness','instrumentalness','time_signature'
                         ,'duration','loudness','valence','danceability','mode','time_signature_confidence','tempo_confidence'
                         ,'key_confidence','mode_confidence']
```

```python
df_feature =  pd.DataFrame(audio_feature_list, columns = audio_feature_header)
df_feature.loc[:,].dropna(axis=0,how='all',inplace=True)
```

```python
df_feature.head(10)
```

```python
df_feature['genres'] = songs_dataset['genres']
```

```python
df_feature.head(10)
```

```python
rap_dance = df_feature.loc[(df_feature['genres'] == 'rap') | (df_feature['genres'] == 'dance')]
```

```python
rap_dance
```

```python
rap_dance.reset_index(drop=True)
```

```python
rap_dance.isnull().any()
```

```python
lables_final = np.array(rap_dance['genres'])
lables_final
```

```python
rap_dance = rap_dance.drop('genres', axis=1) 
```

```python
lables_final
```

```python
rap_dance.isnull().any()
```

```python
rap_dance
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
from sklearn.preprocessing import StandardScaler
standar_scaler = StandardScaler()
rap_dance = standar_scaler.fit_transform(rap_dance)
```

```python
rap_dance
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(rap_dance ,lables_final,test_size = 0.2,random_state = 101)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from nltk import ConfusionMatrix
confusion_matrix(y_test, predictions)
print(ConfusionMatrix(list(y_test), list(predictions)))
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn,rap_dance,lables_final, cv=10, scoring = 'accuracy')
print(scores.mean())
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```
